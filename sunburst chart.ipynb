```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('pepsico_products.csv')
```

```python
df.head()
```

```python
df_new = df[df['Status'] == 'Active']
```

```python
fig = px.sunburst(df_new, path = ["Ownership", "Region Served"], title = "PepsiCo: Time to Expand", color = "Ownership")
fig.update_layout(title_text="PepsiCo: Time to Expand", title_x=0.5)
fig.show()
```
